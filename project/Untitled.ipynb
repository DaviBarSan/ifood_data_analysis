```python
import pandas as pd
```

```python
df = pd.read_csv("./ifood_df.csv")
```

```python
df
```

```python
df_test = df[['education_2n Cycle', 'education_Basic', 'education_Graduation', 'education_Master', 'education_PhD']]
df_test
edu_from_dummies = pd.from_dummies(df_test)
edu_from_dummies
```

```python
df_marital_status = df[['marital_Divorced','marital_Married',	'marital_Single','marital_Together','marital_Widow']]
df_marital_status
marital_from_dummies = pd.from_dummies(df_marital_status)
marital_from_dummies
```

```python
df['maritalStatus'] = marital_from_dummies
```

```python
df['educationLevel'] = edu_from_dummies
```

```python
df.drop(labels=['marital_Divorced','marital_Married',	'marital_Single','marital_Together','marital_Widow'], axis='columns', inplace=True)
df
```

```python
df.drop(labels=['education_2n Cycle', 'education_Basic', 'education_Graduation', 'education_Master', 'education_PhD'], axis='columns', inplace=True)
df
```

```python
df.drop(labels=['Z_CostContact','Z_Revenue'], axis='columns', inplace=True)
df
```
